Notes on the above results:
* Conv + RNN and Conv + Transformer were the best models
* I will continue with CRNN since Transformers had a higher CER on the test set, indicating that it might not be generalizing as well.
* CRNN with 3 RNN layers did not see a large improvement over CRNN with 2 layers (10M params)
* Will be testing Conv filter sizes, hidden sizes, and using LSTM in the future experiments.

some transformer notes:
1. Because the model has a `seq_len` that needs to handle the input. Test set has dim 140633 so context length needs to handle it.
2. Instead of making the max sequence length 140,000 I add a pooling layer to decrease the convolutional output by a factor of `subsample`. Using `subsample = 32` means the `140633` becomes ≈`4394`. I can use `seq_len=4800` to capture this.
3. Because of the pooling layer I had toe recalculate the emissions length in the `_step()` function using that scaling factor.

Top 2 models are CRNN, and Conv + Transformer model.
